```python
# A_ Métro du réseau STAR : localisation des stations
import requests
import pandas as pd

# URL du fichier CSV
csv_url = "https://data.opendatasoft.com/api/explore/v2.1/catalog/datasets/metro_localisation_stations@rennes-metropole/exports/csv?lang=fr&timezone=Europe%2FBerlin&use_labels=true&delimiter=%3B"

# Effectuer une requête GET pour télécharger le fichier CSV
response = requests.get(csv_url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Enregistrer le contenu dans un fichier local
    with open("bus_positions.csv", "wb") as file:
        file.write(response.content)
    print("Fichier CSV téléchargé avec succès.")
    DF1 = pd.read_csv("Métro du réseau STAR_localisation des stations.csv", delimiter=';')  # Charger dans une DataFrame
else:
    print(f"Erreur lors du téléchargement du fichier CSV : {response.status_code}")
```
